```python
import cv2
import os
from matplotlib import pyplot as plt
```

Load Image

```python
img_path='/content/edgedetection1.jpg'
```

```python
img = cv2.imread(img_path)
```

```python
img
```

```python
plt.imshow(img)
```

Edge Detection


```python
#Apply Canny
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
blur =cv2.GaussianBlur(gray, (5,5),0)
canny = cv2.Canny(blur, threshold1=10, threshold2=150)
```

```python
canny
```

```python
plt.imshow(canny)
```

Export Edges

```python
cv2.imwrite('edges.jpg',canny)
```
